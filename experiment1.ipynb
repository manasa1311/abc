```python
import pandas as pd
import numpy as np
#load the datset
dataset=pd.read_csv("/content/age_sal_dataset.csv")
#printing the dataset
dataset
```

```python
#Extraction of independent and dependent variables
X= dataset.iloc[:,:-1].values
Y=dataset.iloc[:,-1].values
print("independent variable(X):",X)
print("Dependent variable (Y)",Y)
```

```python
#dealing with missing data
from sklearn.impute import SimpleImputer
#REPLACING MISSING VALUES USING TRANSFORM METHOD
imp=SimpleImputer(missing_values=np.nan, strategy="mean")
X=imp.fit_transform(X)
Y=Y.reshape(-1,1)
Y=imp.fit_transform(Y)
Y=Y.reshape(-1)
print(X)
print(Y)
```

```python
#covert X to Y to DataFrames
X_df=pd.DataFrame(X, columns=dataset.columns[:-1])
Y_df=pd.DataFrame(Y, columns=[dataset.columns[-1]])
#combine X and Y using pandas.concat
combined_data=pd.concat([X_df, Y_df], axis=1)
print("Combined Data:")
print(combined_data)
```

```python
combined_df=pd.DataFrame(combined_data,columns=list(dataset.columns))
combined_df.to_excel("/content/combined_data.xlsx",index=False)
print("combined dataframe")
print(combined_df.head())
```
